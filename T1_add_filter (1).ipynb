```python
import cv2

img = cv2.imread("image.png")
img = cv2.resize(img, (700, 500))
mode = "color"

while True:
    frame = img.copy()

    if mode == "gray":
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    elif mode == "red":
        frame[:, :, 0] = 0
        frame[:, :, 1] = 0

    elif mode == "green":
        frame[:,:,0] = 0 
        frame[:,:,2] = 0
        
    elif mode == "blue":
        frame[:, :, 1] = 0
        frame[:, :, 2] = 0

    elif mode == "yellow":
        frame[:, :, 0] = 0

    elif mode == "pink":
        frame[:, :, 1] = 0
    elif mode == "cyan":
        frame[:, :, 2] = 0

    elif mode == "orange":
        frame[:,:,0] = 0
        frame[:,:,1] = frame[:,:,1]//2
        
    elif mode == "brown":
        frame[:, :, 0] = frame[:, :, 0] // 10
        frame[:, :, 1] = frame[:, :, 1] // 5

    elif mode == "purple":
        frame[:, :, 1] = 0
        frame[:, :, 0] = frame[:, :, 0] // 2

    elif mode == "light blue":
        frame[:, :, 2] = 0
        frame[:, :, 1] = frame[:, :, 1] // 2

    elif mode == "navy blue":
        frame[:, :, 2] = 0
        frame[:, :, 1] = 0
        frame[:, : ,0] = frame[:, :, 0] // 2
        

    cv2.imshow("Image Filter", frame)

    key = cv2.waitKey(1)

    if key == ord("g"):
        mode = "gray"
        

    elif key == ord("G"):
        mode = "green"
         
    elif key == ord("y"):
        mode = "yellow"

    elif key == ord("r"):
        mode = "red"

    elif key == ord("C"):
        mode = "color"

    elif key == ord("b"):
        mode = "blue"
        
    elif key == ord("p"):
        mode = "pink"    

    elif key == ord("c"):
        mode = "cyan"

    elif key == ord("o"):
        mode = "orange"

    elif key == ord("B"):
        mode = "brown"

    elif key == ord("p"):
        mode = "purple"

    elif key == ord("l"):
        mode = "light blue"

    elif key == ord("n"):
        mode = "navy blue"
        
    elif key == ord("q"):
        break
        



cv2.destroyAllWindows()
```
