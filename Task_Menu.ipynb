```python
import tkinter as tk
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt


def open_task1():
    face_detector = cv2.CascadeClassifier(
        cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
    )

    cap = cv2.VideoCapture(0)

    while True:
        success, frame = cap.read()

        if not success:
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        face = face_detector.detectMultiScale(gray, 1.1, 5)

        for (x, y, w, h) in face:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        cv2.imshow("Face Detection", frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

    cap.release()
    cv2.destroyAllWindows()


def open_task2():
    face_detector = cv2.CascadeClassifier(
        cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
    )

    emoji = cv2.imread("emoji.png")

    if emoji is None:
        print("emoji.png file not found")
        return

    cap = cv2.VideoCapture(0)

    while True:
        success, frame = cap.read()

        if not success:
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        face = face_detector.detectMultiScale(gray, 1.1, 5)

        for (x, y, w, h) in face:
            emoji_resized = cv2.resize(emoji, (w, h))
            frame[y:y+h, x:x+w] = emoji_resized

        cv2.imshow("Emoji Face", frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

    cap.release()
    cv2.destroyAllWindows()


def open_task3():
    img = cv2.imread("image.png")

    if img is None:
        print("image.png file not found")
        return

    img = cv2.resize(img, (700, 500))
    mode = "color"

    while True:
        frame = img.copy()

        if mode == "gray":
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        elif mode == "red":
            frame[:, :, 0] = 0
            frame[:, :, 1] = 0

        elif mode == "green":
            frame[:, :, 0] = 0
            frame[:, :, 2] = 0

        elif mode == "blue":
            frame[:, :, 1] = 0
            frame[:, :, 2] = 0

        elif mode == "yellow":
            frame[:, :, 0] = 0

        elif mode == "pink":
            frame[:, :, 1] = 0

        elif mode == "cyan":
            frame[:, :, 2] = 0

        elif mode == "orange":
            frame[:, :, 0] = 0
            frame[:, :, 1] = frame[:, :, 1] // 2

        elif mode == "brown":
            frame[:, :, 0] = frame[:, :, 0] // 10
            frame[:, :, 1] = frame[:, :, 1] // 5

        elif mode == "purple":
            frame[:, :, 1] = 0
            frame[:, :, 0] = frame[:, :, 0] // 2

        elif mode == "light blue":
            frame[:, :, 2] = 0
            frame[:, :, 1] = frame[:, :, 1] // 2

        elif mode == "navy blue":
            frame[:, :, 2] = 0
            frame[:, :, 1] = 0
            frame[:, :, 0] = frame[:, :, 0] // 2

        cv2.putText(
            frame,
            "g Gray|r Red|G Green|b Blue|y Yellow|p Pink|o Orange|q Quit",
            (20, 30),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.6,
            (255, 255, 255),
            2
        )

        cv2.imshow("Image Filter", frame)

        key = cv2.waitKey(1) & 0xFF

        if key == ord("g"):
            mode = "gray"
        elif key == ord("G"):
            mode = "green"
        elif key == ord("y"):
            mode = "yellow"
        elif key == ord("r"):
            mode = "red"
        elif key == ord("C"):
            mode = "color"
        elif key == ord("b"):
            mode = "blue"
        elif key == ord("p"):
            mode = "pink"
        elif key == ord("c"):
            mode = "cyan"
        elif key == ord("o"):
            mode = "orange"
        elif key == ord("B"):
            mode = "brown"
        elif key == ord("l"):
            mode = "light blue"
        elif key == ord("n"):
            mode = "navy blue"
        elif key == ord("q"):
            break

    cv2.destroyAllWindows()


def open_task4():
    face_detector = cv2.CascadeClassifier(
        cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
    )

    cap = cv2.VideoCapture(0)

    while True:
        success, frame = cap.read()

        if not success:
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        face = face_detector.detectMultiScale(
            gray,
            scaleFactor=1.1,
            minNeighbors=5,
            minSize=(30, 30)
        )

        for (x, y, w, h) in face:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        count_text = f"Faces : {len(face)}"

        cv2.rectangle(frame, (0, 0), (220, 50), (0, 0, 0), -1)

        cv2.putText(
            frame,
            count_text,
            (10, 35),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0, 255, 0),
            2
        )

        cv2.imshow("Faces Counter", frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

    print("Final face count:", len(face))

    cap.release()
    cv2.destroyAllWindows()


def open_task5():
    cap = cv2.VideoCapture(0)

    while True:
        success, frame = cap.read()

        if not success:
            break

        frame = cv2.resize(frame, (400, 300))

        two_video = cv2.hconcat([frame, frame])

        cv2.imshow("Two Part Live Video", two_video)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

    cap.release()
    cv2.destroyAllWindows()


def open_task6():
    cap = cv2.VideoCapture(0)

    while True:
        success, frame = cap.read()

        if not success:
            break

        frame = cv2.flip(frame, 1)

        roi = frame[50:500, 50:500]

        hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

        lower_skin = np.array([0, 30, 60])
        upper_skin = np.array([25, 255, 255])

        mask = cv2.inRange(hsv, lower_skin, upper_skin)

        kernel = np.ones((5, 5), np.uint8)
        mask = cv2.erode(mask, kernel, iterations=1)
        mask = cv2.dilate(mask, kernel, iterations=2)
        mask = cv2.GaussianBlur(mask, (7, 7), 0)

        contours, _ = cv2.findContours(
            mask,
            cv2.RETR_EXTERNAL,
            cv2.CHAIN_APPROX_SIMPLE
        )

        finger_count = 0

        if len(contours) > 0:
            hand = max(contours, key=cv2.contourArea)
            area = cv2.contourArea(hand)

            if area > 7000:
                hull = cv2.convexHull(hand)
                hull_index = cv2.convexHull(hand, returnPoints=False)

                cv2.drawContours(roi, [hand], -1, (0, 255, 0), 2)
                cv2.drawContours(roi, [hull], -1, (255, 0, 0), 2)

                try:
                    defects = cv2.convexityDefects(hand, hull_index)

                    if defects is not None:
                        gaps = 0

                        for i in range(defects.shape[0]):
                            s, e, f, d = defects[i][0]

                            start = tuple(hand[s][0])
                            end = tuple(hand[e][0])
                            far = tuple(hand[f][0])

                            a = math.dist(start, end)
                            b = math.dist(start, far)
                            c = math.dist(end, far)

                            angle = math.degrees(
                                math.acos((b*b + c*c - a*a) / (2*b*c))
                            )

                            if angle <= 90 and d > 10000:
                                gaps += 1
                                cv2.circle(roi, far, 5, (0, 0, 255), -1)

                        finger_count = gaps + 1

                        if finger_count > 5:
                            finger_count = 5

                except:
                    finger_count = 0

        cv2.rectangle(frame, (0, 0), (250, 50), (0, 0, 0), -1)

        cv2.putText(
            frame,
            f"Fingers : {finger_count}",
            (10, 35),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0, 255, 0),
            2
        )

        cv2.rectangle(frame, (50, 50), (500, 500), (255, 0, 0), 3)

        cv2.imshow("Finger Counter", frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

    print("Final finger count:", finger_count)

    cap.release()
    cv2.destroyAllWindows()


def open_task7():
    image = cv2.imread("image.png")

    if image is None:
        print("image.png file not found")
        return

    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    crop_width = int(input("Enter crop width: "))
    crop_height = int(input("Enter crop height: "))

    total_height = image_rgb.shape[0]
    total_width = image_rgb.shape[1]

    center_y = total_height // 2
    center_x = total_width // 2

    start_row = center_y - crop_height // 2
    end_row = center_y + crop_height // 2

    start_col = center_x - crop_width // 2
    end_col = center_x + crop_width // 2

    cropped = image_rgb[start_row:end_row, start_col:end_col]

    fig, axes = plt.subplots(1, 2, figsize=(10, 4))

    axes[0].imshow(image_rgb)
    axes[0].set_title("Original Image")
    axes[0].axis("off")

    axes[1].imshow(cropped)
    axes[1].set_title("Cropped Image")
    axes[1].axis("off")

    plt.show()


def open_task8():
    cap = cv2.VideoCapture(0)

    hand_count = 0
    hand_inside = False

    while True:
        success, frame = cap.read()

        if not success:
            break

        frame = cv2.flip(frame, 1)

        x1, y1 = 200, 100
        x2, y2 = 500, 400

        roi = frame[y1:y2, x1:x2]

        hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

        lower_skin = np.array([0, 30, 60])
        upper_skin = np.array([25, 255, 255])

        mask = cv2.inRange(hsv, lower_skin, upper_skin)

        kernel = np.ones((5, 5), np.uint8)
        mask = cv2.erode(mask, kernel, iterations=1)
        mask = cv2.dilate(mask, kernel, iterations=2)

        contours, _ = cv2.findContours(
            mask,
            cv2.RETR_EXTERNAL,
            cv2.CHAIN_APPROX_SIMPLE
        )

        hand_detected = False

        if len(contours) > 0:
            hand = max(contours, key=cv2.contourArea)
            area = cv2.contourArea(hand)

            if area > 5000:
                hand_detected = True
                cv2.drawContours(roi, [hand], -1, (0, 255, 0), 2)

        if hand_detected and hand_inside == False:
            hand_count += 1
            hand_inside = True

        if hand_detected == False:
            hand_inside = False

        count_text = f"Hand Count : {hand_count}"

        cv2.rectangle(frame, (0, 0), (330, 50), (0, 0, 0), -1)

        cv2.putText(
            frame,
            count_text,
            (10, 35),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0, 255, 0),
            2
        )

        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 3)

        cv2.imshow("Hand Movement Counter", frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

    print("Final hand movement count:", hand_count)

    cap.release()
    cv2.destroyAllWindows()


window = tk.Tk()
window.title("Computer Vision Dashboard")
window.geometry("420x570")
window.config(bg="lightgray")

title = tk.Label(
    window,
    text="Computer Vision Tasks",
    font=("Arial", 20, "bold"),
    bg="lightgray"
)
title.pack(pady=20)

btn1 = tk.Button(window, text="Launch Camera", width=28, height=2, command=open_task1)
btn1.pack(pady=5)

btn2 = tk.Button(window, text="Emoji on Face", width=28, height=2, command=open_task2)
btn2.pack(pady=5)

btn3 = tk.Button(window, text="Image Filters", width=28, height=2, command=open_task3)
btn3.pack(pady=5)

btn4 = tk.Button(window, text="Face Detect Counter", width=28, height=2, command=open_task4)
btn4.pack(pady=5)

btn5 = tk.Button(window, text="Multiple Video Divide", width=28, height=2, command=open_task5)
btn5.pack(pady=5)

btn6 = tk.Button(window, text="Finger Detection", width=28, height=2, command=open_task6)
btn6.pack(pady=5)

btn7 = tk.Button(window, text="Image Crop", width=28, height=2, command=open_task7)
btn7.pack(pady=5)

btn8 = tk.Button(window, text="Hand Movement Detection", width=28, height=2, command=open_task8)
btn8.pack(pady=5)

window.mainloop()
```
