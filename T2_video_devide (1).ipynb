```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    succes, frame = cap.read()

    if not succes:
        break

    frame = cv2.resize(frame, (400, 300))

    two_video = cv2.hconcat([frame, frame])

    cv2.imshow("Two Part Live Video", two_video)

    if cv2.waitKey(1) == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
